```python
# import necessary modules
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# read in each freq file as a df Darwin
ntIIntIIIa = pd.read_table('ntII-ntIIIa_filtered.frq', sep="\t")
ntIIntIIIa.head()
```

```python
ntIIIa_comp=[]
for i in range(len(ntIIntIIIa)):
    if ntIIntIIIa['ALLELE_1_ntII'].iloc[i] == ntIIntIIIa['ALLELE_1_ntIIIa'].iloc[i]:
        ntIIIa_comp.append(ntIIntIIIa['FREQ_1_ntIIIa'].iloc[i])
    elif ntIIntIIIa['ALLELE_1_ntII'].iloc[i] == ntIIntIIIa['ALLELE_2_ntIIIa'].iloc[i]:
        ntIIIa_comp.append(ntIIntIIIa['FREQ_2_ntIIIa'].iloc[i])
    elif ntIIntIIIa['ALLELE_1_ntII'].iloc[i] == ntIIntIIIa['ALLELE_3_ntIIIa'].iloc[i]:
        ntIIIa_comp.append(ntIIntIIIa['FREQ_3_ntIIIa'].iloc[i])
    else:
        ntIIIa_comp.append('NA')
ntIIntIIIa['ntIIIa_comp']=ntIIIa_comp
ntIIntIIIa.head()
```

```python
# remove scaffolds
patternDel="Scaffold"
filter=ntIIntIIIa['CHROM_ntII'].str.contains(patternDel)
ntIIntIIIa=ntIIntIIIa[~filter]
ntIIntIIIa.head()
```

```python
# calculate dxy
#dxy =(p * (1-q)) + (q * (1-p))
ntIIntIIIa['dxy']=(ntIIntIIIa['FREQ_1_ntII']*(1-ntIIntIIIa['ntIIIa_comp']))+(ntIIntIIIa['ntIIIa_comp']*(1-ntIIntIIIa['FREQ_1_ntII']))
ntIIntIIIa.head()
```

```python
pd.DataFrame.to_csv(ntIIntIIIa, 'ntII-ntIIIa_filtered_dxy.frq', sep="\t", index=False)
```
